```python
import firebase_admin
from firebase_admin import credentials, db, initialize_app

cred = credentials.Certificate("bitcoin-realtime-database-firebase-adminsdk-g4fqg-497fdddf36.json")
app = firebase_admin.initialize_app(cred)

ref = db.reference("/",app=app, url='https://bitcoin-realtime-database-default-rtdb.firebaseio.com/')
ref
```

```python
data = ref.get()
data
```

```python
import pandas as pd
import json
```

```python
data = json.dumps(data)
type(data)
dt = json.loads(data)
```

```python
bt_data = pd.read_json(dt, orient='index')
bt_data
```
